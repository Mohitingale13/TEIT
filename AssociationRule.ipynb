```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
df=pd.read_csv('Market_Basket.csv', header=None)
```

```python
df
```

```python
df.head(10)
```

```python
plt.figure(figsize=(20,10))
df[0].value_counts().plot.bar()
plt.title('Bar Model', fontsize=30)
plt.grid()
#plt.show()
```

```python
products=[]
for i in range(0,7501):
    products.append([str(df.values[i,j]) for j in range(20)])
```

```python
print(products[1])
```

```python
model=apriori(products, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2, max_length=2)
```

```python
results= list(model)
for i in results:
    print('\n')
    print(i)
    print('***********')
```

```python
for item in results:
    pair=item[0]
    items=[x for x in pair]
    print("Rule: " +items[0] +" --> " +items[1])
    print("Support: " +str(item[1]))
    print("Confidence: " +str(item[2][0][2]))
    print("Lift: "+str(item[2][0][3]))
    print('============================================')
```
