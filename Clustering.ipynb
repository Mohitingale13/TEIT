```python
from sklearn.cluster import KMeans
```

```python
wcss_list=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss_list)
#below data is not mandatory without above data you plot graph 
plt.title('Elbow graph')
plt.xlabel('No of clusters')
plt.ylabel('wcss list')
plt.show()
```

```python
len(wcss_list)
```

```python
kmeans=KMeans(n_clusters=5, init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
print("Centroids: ", kmeans.cluster_centers_)
print("Predicted clusters for data points")
y_predict
```

```python
plt.scatter(x.iloc[y_predict == 0,0], x.iloc[y_predict == 0,1] , s=100, c='pink')
plt.scatter(x.iloc[y_predict == 1,0], x.iloc[y_predict == 1,1] , s=100, c='blue')
plt.scatter(x.iloc[y_predict == 2,0], x.iloc[y_predict == 2,1] , s=100, c='orange')
plt.scatter(x.iloc[y_predict == 3,0], x.iloc[y_predict == 3,1] , s=100, c='green')
plt.scatter(x.iloc[y_predict == 4,0], x.iloc[y_predict == 4,1] , s=100, c='yellow')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200 , c='black')

plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
df['cluster']=y_predict
df
```
